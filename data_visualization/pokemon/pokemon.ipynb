# 2 - Graphs

We can begin by plotting a simple static graph, to check the amount of Pokemons given their Type 1. We will use Seaborn library for this.

```python
sns.factorplot(
    x='type1', 
    data=pokemon_df,
    size=10,
    kind='count',
).set_axis_labels('Type 1', 'Amount of Pokemons')

plt.show()
```

We will now plot interactive graphs, and we will check the distribution of Pokemons, given their attack and defense. We will also see a 3rd dimension which is type1.

```python
fig = px.scatter(pokemon_df, 
                 x="attack", 
                 y="defense", 
                 color="type1", 
                 marginal_y="violin", 
                 marginal_x="box",
                 hover_name="name",
                 #trendline='ols',
                 title="Attack, defense and type of Pokemons")
fig.show()
```

```python
pokemon_df['type2'].isnull().sum() # to check that a Pokemon can only have a type 1 and no type 2
```

```python
fig = px.parallel_categories(pokemon_df,
                             dimensions=['type1', 'type2', 'is_legendary'],
                             title="Number of pokemons by type 1, type 2 and legendarity")
fig.show()
```

```python
fig = px.violin(pokemon_df, 
                y="speed",  
                color="generation", 
                box=True, 
                points="all", 
                hover_data=pokemon_df.columns,
                title='Distribution of Pokemon speed per generation')
fig.show()
```

```python
import plotly.graph_objs as go
```

```python
pokemon_df.columns
```

```python
x = pokemon_df[pokemon_df["name"] == "Charizard"]
data = [go.Scatterpolar(
  r = [x['hp'].values[0],
       x['attack'].values[0],
       x['defense'].values[0],
       x['sp_attack'].values[0],
       x['sp_defense'].values[0],
       x['speed'].values[0],
       x['hp'].values[0]],
  theta = ['hp','attack','defense','sp_attack','sp_defense','speed','hp'],
  fill = 'toself',
     line =  dict(
            color = 'orange'
        )
)]

layout = go.Layout(
  polar = dict(
    radialaxis = dict(
      visible = True,
      range = [0, 200]
    )
  ),
  showlegend = False,
  title = "{} stats distribution".format(x['name'].values[0])
)
fig = go.Figure(data=data, layout=layout)
fig.layout.images = [dict(
        source="https://rawgit.com/guramera/images/master/Charizard.png",
        xref="paper", yref="paper",
        x=0.95, y=0.3,
        sizex=0.6, sizey=0.6,
        xanchor="center", yanchor="bottom"
      )]

fig.show()
```
